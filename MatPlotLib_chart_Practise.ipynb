```
import matplotlib.pyplot as plt
```

```
# Create Pie Chart, plt.pie() for pie chart and autopct= is for percent shows
# >1f shwos to display till 1 value after decimal
# Here we can explode the bigger chart USA 
# We can define the color also as per requirement
plt.figure(figsize=(5,5))
labels = ["USA","INDIA","GERMANY","AUSTRALIA","NEWZEALAND"]
values = [25, 10, 15, 13, 7]
explode = [0.05,0,0,0,0]
colors = ["c","g","r","y","b"]
plt.pie(values,labels=labels, autopct="%.1f%%", explode = explode, colors=colors)
plt.show()
```

```
# Creating Bar chart compairision of mobile price
# We can change the width of the bar chart
# Make more viualize the x axis with corresponding names
plt.figure(figsize=(7,5))
names = ["Lava","Micromax","Samsung","Xiomi"]

price_2018 = [3000, 3500, 6000, 3200]
price_2019 = [2500, 4000, 5000, 4500]

position = [0,1,2,3]
position2 = [0.3,1.3,2.3,3.3]
position_names = [.15, 1.15, 2.15, 3.15] 
plt.bar(position, price_2018, width=0.3, color = 'r')
plt.bar(position2, price_2019, width=0.3)

plt.title("Mobile price of 2018 and 2019 ")
plt.xticks(position_names, names)
plt.show()
```
